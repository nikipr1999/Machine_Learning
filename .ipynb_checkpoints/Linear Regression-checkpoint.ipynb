# Linear Regression Model

```python
home.info()
```

```python
X=home[['bedrooms','bathrooms','sqft_living15','sqft_lot','floors','sqft_above','sqft_lot15','yr_built','condition','zipcode']]
y=home['price']
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.35)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
prediction[1]    #actual value is 534k
```

```python
plt.figure(figsize=(10,8))
plt.scatter(y_test,prediction)
```

```python
X.shape
```

```python
model.coef_
# these are respective coeffecient for our models equation
```

```python
model_coef=pd.DataFrame(model.coef_,X.columns)
```

```python
model_coef
```
