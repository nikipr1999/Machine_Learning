```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
import numpy as np
```

```python
num=np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num,color='red')
```

```python
label_dist=pd.Series(num,name='variable x')
```

```python
sns.distplot(label_dist)
```

```python
#only KDE is drawn 
sns.distplot(label_dist,hist=False)
```

```python
sns.distplot(label_dist,vertical=True)
```

```python
sns.distplot(label_dist,hist=False,rug=True,color='orange')
```

```python
sns.distplot(label_dist,hist=False,rug=True,vertical=True,color='orange')
```
