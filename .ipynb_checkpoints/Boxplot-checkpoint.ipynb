```python
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
tips=pd.read_csv('tips')
```

```python
tips.head()
```

```python
sns.boxplot(x=tips.size)
```

```python
sns.boxplot(x=tips.total_bill)
```

```python
sns.boxplot(x=tips.sex,y=tips.total_bill)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',order=['Thurs','Fri','Sat','Sun'],palette='winter_r',orient='vertical')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',order=['Thurs','Fri','Sat','Sun'],palette='spring',orient='vertical')
# palette changed to spring
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips)
# if we type in same cell then plots will be in same graph
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```
