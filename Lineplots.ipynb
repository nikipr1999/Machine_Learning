```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path='Canada.xlsx'
```

```python
df=pd.read_excel(path,skiprows=range(20))
```

```python
df.head()
```

```python
df.set_index(df.RegName,inplace=True,drop=True,append=False)
```

```python
df.head()
```

```python
years=list(map(str,range(1980,2014)))
df.loc['AreaName',years].plot(kind='line')
plt.xlabel('Years')
plt.ylabel('Number of immigration')
plt.title('Immigration from Haiti')
```
